```python
query = f"""
SELECT
  dept_no,
  employee_dept_name as dept_name,
  employment_status,
  count(emp_no) as employee_count,
  employee_dept_start_date as from_date,
  employee_dept_end_date as to_date
FROM employee_catalog.gold.employee_details
GROUP BY
  1, 2, 3, 5, 6
ORDER BY
  1, 2
"""
```

```python
alltime_dept_trend_df = spark.sql(query)
display(alltime_dept_trend_df)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS employee_catalog.gold")
```

```python
alltime_dept_trend_df.write \
    .option("mergeSchema", True) \
    .mode("overwrite") \
    .saveAsTable("employee_catalog.gold.alltime_dept_trend")
```

```python
alltime_dept_trend_df.write \
    .format("parquet") \
    .mode("overwrite") \
    .save("mnt/gold/alltime_dept_trend")
```
