```python
query = f"""

SELECT 
    s.emp_no,
    e.full_name,
    s.salary,
    LAG(s.salary) OVER (PARTITION BY s.emp_no ORDER BY s.from_date) AS previous_salary,
    ROUND((s.salary - LAG(s.salary) OVER (PARTITION BY s.emp_no ORDER BY s.from_date)) / NULLIF(LAG(s.salary) OVER (PARTITION BY s.emp_no ORDER BY s.from_date), 0) * 100, 2) AS salary_growth_percentage,
    s.from_date AS salary_start_date,
    s.to_date AS salary_end_date
FROM employee_catalog.silver.salaries s
LEFT JOIN employee_catalog.silver.employees e 
ON s.emp_no = e.emp_no
ORDER BY
    1

"""
```

```python
salary_growth_df = spark.sql(query)
display(salary_growth_df)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS employee_catalog.gold")
```

```python
salary_growth_df.write \
    .option("mergeSchema", True) \
    .mode("overwrite") \
    .saveAsTable("employee_catalog.gold.salary_growth")
```

```python
salary_growth_df.write \
    .format("parquet") \
    .mode("overwrite") \
    .save("mnt/gold/salary_growth")
```
